## Analyse des données:

### Chiffre d’affaires par produit, ainsi que le total du chiffre d’affaires réalisé en ligne.

* Selection de la table et des colonnes:

```python
df_web_produit=df_web.loc[:,['id_web','total_sales', 'post_name']]
df_web_produit
```

* Effectuons une jointure entre les tables web et liaison:

```python
df_web_produit_liaison=pd.merge(df_web_produit, df_liaison, how='left', on='id_web')
df_web_produit_liaison
```

*Nous avons alors ici une table qui regroupe id_web et product_id.*

* Effectuons ensuite une jointure entre cette meme table et la table erp:

```python
df_erp_web=pd.merge(df_erp, df_web_produit_liaison ,how='right', on='product_id')
df_erp_web
```

* Classement des produits les plus vendus:

```python
df_erp_web_top_vente = df_erp_web.sort_values(by='total_sales', ascending=False).head (20)
df_erp_web_top_vente
```

*Nous avons ici la somme des ventes par produits.*

```python
plt.subplots(figsize = (20,10))
height = df_erp_web_top_vente['total_sales']
bars = df_erp_web_top_vente['post_name']
x_pos = np.arange(len(bars))
plt.bar(x_pos, height)
plt.xticks(x_pos, bars, rotation=90)
plt.subplots_adjust(bottom=0.4, top=0.99)   
plt.xlabel('Nom des produits',fontsize = 15)
plt.ylabel("Nombres de ventes par produit",fontsize = 15) 
plt.title('Top 20 meilleurs ventes',fontsize = 20 )
plt.show()
```

* Ajoutons à cette table le chiffre d'affaires par produit:

```python
df_erp_web['CA_produit']=df_erp_web['price']*df_erp_web['total_sales']
df_erp_web
```

* Effectuons un classement sur les 20 produits les plus rentables:

```python
df_erp_web_top20=df_erp_web.sort_values(by='CA_produit', ascending=False).head(20)
df_erp_web_top20
```

```python
plt.subplots(figsize = (20,10))
height = df_erp_web_top20['CA_produit']
bars = df_erp_web_top20['post_name']
x_pos = np.arange(len(bars))
plt.bar(x_pos, height)
plt.xticks(x_pos, bars, rotation=90)
plt.subplots_adjust(bottom=0.4, top=0.99)   
plt.xlabel('Nom des produits',fontsize = 15)
plt.ylabel("Chiffre d'affaires par produit",fontsize = 15) 
plt.title('Top 20 des produits les plus rentables',fontsize = 20 )
plt.show()
```

* Calculons le chiffre d'affaires total des produits:

* Observons alors s'il existe des valeurs aberrantes:

```python
df = df_erp_web
df

sns.boxplot( y=df["price"] )
plt.show()
```

*Nous observons quelques valeurs au dessus des prix habituels.*

* Cherchons maintenant les valeurs aberrantes grace à la règle interquartille:

Pour détecter les valeurs aberrantes, nous utiliserons l'intervalle interquartile.

Premier quartille:

```python
Q1 = df_erp_web['price'].quantile(.25)
Q1
```

Troisième quartille:

```python
Q3 = df_erp_web['price'].quantile(.75)
Q3
```

L'écart interquatile donne une idée plus juste de la dispersion des données:
q3-q1

```python
EIQ=Q3-Q1
EIQ
```

*Nous avons donc une valeur plus représentative des prix qui est ici de 28 euros.*

* La prochaine étape sera de trouver la valeur au dela de laquelle les autres valeurs seront aberrantes:

Pour cela nous allons utiliser la formule suivante:

```python
limite_valeurs_aberrantes = 1.5*EIQ+Q3
limite_valeurs_aberrantes
```

Nous avons donc toutes les valeurs au dessus de 84€ qui seront aberrantes.

* Affichons alors le nombre de valeurs aberrantes:

```python
val_abb = np.where(df_erp_web['price']>=limite_valeurs_aberrantes)
val_abb
```

```python
nPval_abb = np.array(val_abb)
nPval_abb.size
```

*Nous avons ici 32 valeurs qui sont au dessus des prix habituels.*

* Voyons donc quelles sont ces valeurs:

```python
df_erp_web_prix_top32=df_erp_web.sort_values(by='price', ascending=False).head(32)
df_erp_web_prix_top32
```

* Calculons le chiffre d'affaires que représente ces 32 produits:

```python
total_ca_top32=df_erp_web_prix_top32['CA_produit'].sum()
total_ca_top32
```

*Nous obtenons un total de 6970€ pour 32 produits.*

* comparons alors cette valeur au chiffre d'affaires total réalisé:

```python
total_ca = ['total_ca_top32','total_ca_produit'] 
  
data = [total_ca_top32, total_ca_produit] 
  
colors = ( "red", "royalblue") 
  
wp = { 'linewidth' : 1, 'edgecolor' : "green" } 
  
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n{:d}€".format(pct, absolute) 
  
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data),   
                                  labels = total_ca, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
  
ax.legend(wedges, total_ca, 
          title ="total_ca", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Proportion du CA des top price sur le CA total") 
  
plt.show() 
```

*Nous obtenons 9% des ventes totales avec 32 produits.*

* Voyons ce que représente cette quantité sur la totalité des produits:

```python
df_erp_web_prix_top32['product_id'].count()
```

```python
df_erp_web['product_id'].count()
```

```python
rapport_top32_total = ['top_32','nombre_total_produits'] 
  
data = [df_erp_web_prix_top32['product_id'].count(), df_erp_web['product_id'].count()] 
  
colors = ( "red", "royalblue") 
  
wp = { 'linewidth' : 1, 'edgecolor' : "green" } 
  
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n{:d}".format(pct, absolute) 
  
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data),   
                                  labels = rapport_top32_total, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
  
ax.legend(wedges, rapport_top32_total, 
          title ="Nombre-ventes", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Proportion des ventes top_32 sur le nombre total de produits") 
  
plt.show() 
```

*Avec seulement 4,3% du stock, nous avons réalisé 9% du chiffre d'affaires.*
